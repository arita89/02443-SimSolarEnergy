Getting the data from https://quickstats.censusdata.abs.gov.au/census_services/getproduct/census/2016/quickstat/LGA70200 for the population of Alice Springs. Copying it into an excel file, now importing all the required packages and then importing the file to python. 

```python
import os
import numpy as np
import urllib.request
import pandas as pd

df = pd.read_excel (r'Data_population.xlsx', dtype = {'Age': np.str, 'population': np.int32 , 'percentage': np.float})
print (df)
```

Now rearranging the groups. So they fit our model. 

```python
b_r = df.iloc[2,2] + df.iloc[3,2] + df.iloc[0,2] + df.iloc[1,2]
b_p = df.iloc[2,1] + df.iloc[3,1] + df.iloc[0,1] + df.iloc[1,1]

c_r = df.iloc[4,2] + df.iloc[5,2]
c_p = df.iloc[4,1] + df.iloc[5,1]

d_r = df.iloc[6,2] + df.iloc[7,2] + df.iloc[8,2]
d_p = df.iloc[6,1] + df.iloc[7,1] + df.iloc[8,1]

e_r = df.iloc[9,2] + df.iloc[10,2] + df.iloc[11,2]
e_p = df.iloc[9,1] + df.iloc[10,1] + df.iloc[11,1]

f_r = df.iloc[12,2] + df.iloc[13,2] + df.iloc[14,2]
f_p = df.iloc[12,1] + df.iloc[13,1] + df.iloc[14,1]

g_r = df.iloc[15,2] + df.iloc[16,2] + df.iloc[17,2]
g_p = df.iloc[15,1] + df.iloc[16,1] + df.iloc[17,1]

data = {'Age Range':['0-19', '20-29', '30-44', '45-59', '60-74' , '75-<'],
       'Population': [b_p , c_p , d_p , e_p , f_p , g_p],
       'Percentage': [b_r , c_r , d_r , e_r , f_r , g_r]}
final_data_frame_population = pd.DataFrame(data)
print(final_data_frame)
```
