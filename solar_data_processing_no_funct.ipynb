Solar Data collection and processing

Solar energy data for the area of Alice Springs is retrieved from http://reg.bom.gov.au/climate/reg/oneminsolar/. Since the activity data is from the year 2014, the extracted data for the solar energy will also be from 2014. 

The collected solar data is per each minute of that year. The data is required to be per 10 minutes.

```python
def addTimeColumn(df):
    time_list = []
    
    #starting point of the time series, just change it as appropriate
    initial_date ='01012014 00:00:00' 
    initial_DateTime = datetime.strptime(initial_date,'%d%m%Y %H:%M:%S') #(2014, 1, 1, 0, 0)
    currentDateTime = initial_DateTime

    counter = 0
    while counter <= len(df.Time):
        next_DateTime= initial_DateTime + timedelta(minutes = 10*counter)
        #next_DateTime= next_DateTime.strftime('%d-%m-%Y %H:%M:%S')
        time_list.append(next_DateTime)
        currentDateTime = next_DateTime
        counter +=1
        
    time_list = pd.Series(time_list, name = 'Date_Time')
    new_column_df = time_list.to_frame()
    df = pd.concat([df, new_column_df],axis=1)
    
    return df

def createTimeStates (df_test):
    time_period = []
    for i in range((len(df_test.Date_Time))):
        tp = (round(df_test.Date_Time[i].minute / 10) + df_test.Date_Time[i].hour * 6)
        #print(tp)
        time_period.append(tp)

    time_period = pd.Series(time_period, name = 'time_period')
    new_column_df = time_period.to_frame()    

    #concat on axis = 1 --> we add one column
    df_test = pd.concat([df_test, new_column_df],axis=1)
    return df_test

def addDayOfWeek(df_test):
    df_test['day_of_week'] = df_test['Date_Time'].dt.day_name()
    return df_test

def getSeason(date_time):
    #print (type(date_time))
    month = date_time.month
    season = [(month%12 + 3)//3]
    #print ((season[0]))
    if (season[0] == 1):
        season_verbose = 'winter'
    elif (season[0] == 2):
        season_verbose = 'spring'
    elif (season[0] == 3):
        season_verbose = 'summer'
    else:
        season_verbose = 'autumn'
    return season_verbose 
    #return season 

def addSeason(df_test):
    local = df_test.copy()
    local['Season']= local['Date_Time'].apply(getSeason)
    return local

#need to convert date time formatting and split it 
def convertandsplitdatetime(df_test): 
    index = df_test.columns.get_loc("Date_Time")
    df_test['Date_Time'] = pd.to_datetime(df_test['Date_Time'], format='%Y-%m-%dT%H:%M:%SZ')
    new_date = [d.date() for d in df_test['Date_Time']]
    new_time = [d.time() for d in df_test['Date_Time']]
    #df_test = df_test.drop(columns = ['t_start'])
    df_test.insert (index, "Date_full", new_date)
    df_test.insert (index+1, "Time_full", new_time)
    return df_test

print()
```
